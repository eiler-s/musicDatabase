### Comments on Interesting Queries in Tables

Language:

This query shows the difference between column 005 (Native speakers of Spanish older than 5 with low proficiency in English) and column 004 (Native speakers of Spanish older than 5 who speak English very well). A large difference in proficiency in English within a community may indicate that there are more recent immigrants to that community than there are established bilingual households.

```python
%%bigquery
select State, GEOID, AreaName, B16001_005, B16001_004 from us_census_staging.Language
where (B16001_005 - B16001_004) > 5000 order by (B16001_005 - B16001_004) asc
```

Race:

This query shows several columns (001- total population, 002- white population, 003-African American population, 005-Asian population). The query is looking for rows where the white population is smaller than both the Asian and African American populations. I thought this might be a good measure of how diverse an area is, however this query only returned areas where the overall population (001) is very small to start with, however, it will be interesting to compare race data to other tables like earnings to get a more complete picture of diversity within the country.

```python
%%bigquery
select State, GEOID, AreaName, B02001_001, B02001_002, B02001_003, 
B02001_005 from us_census_staging.Race where B02001_002 < B02001_003 and B02001_002 < B02001_005
order by B02001_001
```

Education:

B15002_001 - SEX BY EDUCATIONAL ATTAINMENT FOR THE POPULATION 25 YEARS AND OVER for Population 25 Years And Over

B15002_015 - SEX BY EDUCATIONAL ATTAINMENT FOR THE POPULATION 25 YEARS AND OVER for Population 25 Years And Over% Male:% Bachelor's degree

B15002_032 - SEX BY EDUCATIONAL ATTAINMENT FOR THE POPULATION 25 YEARS AND OVER for Population 25 Years And Over% Female:% Bachelor's degree

This query looks for places where more women than men have bachelors degrees compared to the total population. The areas are limited to places where there are more than 2000 people in the area because after an initial exploration it is clear that in some areas there are 0 people with bachelors degrees.

```python
%%bigquery
select State, GEOID, AreaName, B15002_001, B15002_015, B15002_032
from us_census_staging.Education
where B15002_032 > B15002_015 and B15002_001 > 2000
order by B15002_001
```

Age & Sex:


B01002A_001 - median age of while only population
B01002B_001 - median age of black only population
B01002G_001 - median age of two or more races population

This query is meant to identify counties in which the average age of mixed
raced citizens exceeds the average of other citizens of only white or black
races. This would indicate that the county is well mixed racially as racial
mixing would have had to occur earlier in the past rather than the present
day. This would make these areas targets for studying the acceptance of
racial difference within the local cultures. The age of mixed race
individuals is compared to those of whites and blacks because the latter
two groups have historically been the most numerous populations of the
southern united states.

```python
%%bigquery
select State, GEOID, AreaName, B01002A_001, B01002B_001, B01002G_001 from
us_census_staging.AgeSex
where B01002G_001 > B01002A_001
or B01002G_001 > B01002B_001
order by B01002_001 desc
```

Earnings:
B20001_001 - SEX BY EARNINGS IN THE PAST 12 MONTHS (IN 2015 INFLATION-ADJUSTED DOLLARS) FOR THE POPULATION 16 YEARS AND OVER WITH EARNINGS IN THE PAST 12 MONTHS for Population 16 Years And Over With Earnings

B20001_002 - SEX BY EARNINGS IN THE PAST 12 MONTHS (IN 2015 INFLATION-ADJUSTED DOLLARS) FOR THE POPULATION 16 YEARS AND OVER WITH EARNINGS IN THE PAST 12 MONTHS for Population 16 Years And Over With Earnings% Male:

B20001_023 - SEX BY EARNINGS IN THE PAST 12 MONTHS (IN 2015 INFLATION-ADJUSTED DOLLARS) FOR THE POPULATION 16 YEARS AND OVER WITH EARNINGS IN THE PAST 12 MONTHS for Population 16 Years And Over With Earnings% Female:

This query searches the data for areas where 

women made more money than men. It is surprising to me that so many areas met this characteristic. This table could be further combined with race data or age and sex data to further determine which groups make more money.

```python
%%bigquery
select State, GEOID, AreaName, B20001_001, B20001_002, B20001_023 from us_census_staging.Earnings
where B20001_023 > B20001_002
order by B20001_001 desc
```
