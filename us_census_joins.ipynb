We will show three queries in the following section that focus on the educational attainment of the Hispanic population in the southern United States. The first two queries are inner joins with the race data table. The first joins with the education table while the second uses the age and sex data table. The third join will be an outer join of the two tables generated by the previous two queries. Note that Hispanic is a cultural ethnicity and not a race. As a result, information about this group is difficult to derive from race data alone, hence our rational for joining other data tables with the general population data contained in the race table.

The first join is an inner join between the race and education data tables, using GEOID as the shared attribute. This query identifies areas where there is a white majority in the population. The achievement of bachelors degrees by the white population is compared to respective bachelors degrees achieved by the male and female Hispanic populations. The results are ordered by the disparity in bachelors degree achievement between the overall white population and all Hispanics specifically.

- GEOID - Geographical identification code of an area
- AreaName - English names of areas corresponding with GEOIDs
- B02001_001 - Total population of an area
- B02001_002 - Total white population in an area
- C15002A_001 - Population of white people holding bachelors degrees in an area
- C15002I_002 - Population of Hispanic men holding bachelors degrees in an area
- C15002I_007 - Population of Hispanic females holding bachelors degrees in an area

```python
%%bigquery
select r.GEOID, r.AreaName, r.B02001_001,
    edu.C15002A_001, edu.C15002I_002, edu.C15002I_007
from us_census_staging.Race as r inner join us_census_staging.Education as edu
on r.GEOID = edu.GEOID
where r.B02001_002 > r.B02001_001 / 2
order by C15002A_001 - (edu.C15002I_002 + edu.C15002I_007) desc
limit 12
```

The second join is an inner join between the race and age-sex data tables, using GEOID as the shared attribute. This query identifies areas where there are Hispanic populations, as is indicated by a non-null median age for the male and female groups of this ethnicity. The results are ordered to show areas where there are older Hispanic males than Hispanic females. The rational for this is that regions with older male Hispanics than females may indicate a larger Hispanic male population. This may indicate that there are more Hispanic men without wives or children, as is common in groups of migrant workers.

- GEOID - Geographical identification code of an area
- AreaName - English names of areas corresponding with GEOIDs
- B02001_001 - Total population of an area
- B01002_001 - Average age of the total population in an area
- B01002I_002 - Average age of a Hispanic male in an area
- B01002I_003 - Average age of a Hispanic female in an area

```python
%%bigquery
select r.GEOID, r.AreaName, r.B02001_001,
    a.B01002_001, a.B01002I_002, a.B01002I_003
from us_census_staging.Race as r inner join us_census_staging.AgeSex as a
on r.GEOID = a.GEOID
where a.B01002I_002 != '.' and a.B01002I_003 != '.'
order by (cast(a.B01002I_002 as numeric) - cast(a.B01002I_003 as numeric)) desc
limit 12
```

The final join is an outer join between the tables produced by the previous two queries. The join is designed to retain the areas where there are large differences in age between male and female Hispanics. More precisely, the GEOIDs of the second query, areas of age difference, will remain while only corresponding GEOIDs of the first query, areas of disparity in educational achievement, will also be retained. Other GEOIDs where there are disparities in educational achievement, but not differences in Hispanic ages, will not be included in the outer join. Only areas where there are at least some people who attain bachelors degrees in the white population will be selected. Furthermore, the results will be ordered by size of the total population in an area.

- GEOID - Geographical identification code of an area
- AreaName - English names of areas corresponding with GEOIDs
- B02001_001 - Total population of an area
- B01002_001 - Average age of the total population in an area
- B01002I_002 - Average age of a Hispanic male in an area
- B01002I_003 - Average age of a Hispanic female in an area
- C15002A_001 - Population of white people holding bachelors degrees in an area
- C15002I_002 - Population of Hispanic men holding bachelors degrees in an area
- C15002I_007 - Population of Hispanic females holding bachelors degrees in an area

```python
%%bigquery
select r.GEOID, r.AreaName, r.B02001_001,
    a.B01002_001, edu.C15002A_001,
    a.B01002I_002, edu.C15002I_002,
    a.B01002I_003, edu.C15002I_007
from us_census_staging.Race as r
  inner join us_census_staging.AgeSex as a on r.GEOID = a.GEOID
  left outer join us_census_staging.Education as edu on r.GEOID = edu.GEOID
where edu.C15002A_001 is not null
order by r.B02001_001 desc
limit 12
```
